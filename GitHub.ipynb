# Git Installation

## Windows
1. Visit the official Git website: https://git-scm.com/downloads
2. Download the latest version of Git for Windows
3. Run the installer and follow the instructions

Note: During the installation process, you can choose the default settings or customize them as per your requirements.

## Linux (Ubuntu/Debian)
1. Open the terminal
2. Update the package index: `sudo apt-get update`
3. Install Git: `sudo apt-get install git`

## macOS
1. Visit the official Git website: https://git-scm.com/downloads
2. Download the latest version of Git for macOS
3. Run the installer and follow the instructions

Alternatively, you can install Git using Homebrew:
1. Install Homebrew if you haven't already: `/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"`
2. Install Git: `brew install git`

# GitHub Registration

1. Visit the GitHub website: https://github.com
2. Click on the "Sign up" button in the top-right corner
3. Enter your desired username, email address, and password
4. Complete the remaining steps, such as choosing your plan (free or paid) and verifying your email address

# GitHub Configuration

## Windows
1. Open Git Bash
2. Set your username: `git config --global user.name "Your Name"`
3. Set your email: `git config --global user.email "your@email.com"`

## Linux/macOS
1. Open the terminal
2. Set your username: `git config --global user.name "Your Name"`
3. Set your email: `git config --global user.email "your@email.com"`

# Creating a Repository and Linking with GitHub

1. Log in to your GitHub account
2. Click on the "+" icon in the top-right corner and select "New repository"
3. Enter a name for your repository and choose the desired settings (e.g., public or private, initialize with a README file)
4. Click "Create repository"
5. Copy the repository URL (e.g., https://github.com/your-username/your-repo.git)

## Windows
6. Open Git Bash
7. Navigate to the desired location for your local repository: `cd /path/to/your/directory`
8. Initialize a new Git repository: `git init`
9. Link your local repository to the GitHub repository: `git remote add origin https://github.com/your-username/your-repo.git`

## Linux/macOS
6. Open the terminal
7. Navigate to the desired location for your local repository: `cd /path/to/your/directory`
8. Initialize a new Git repository: `git init`
9. Link your local repository to the GitHub repository: `git remote add origin https://github.com/your-username/your-repo.git`

# First Commit

1. Create a new file called `example.py` in your local repository
2. Add some example Python code to the file
3. Stage the changes: `git add example.py`
4. Commit the changes: `git commit -m "Initial commit with example.py"`
5. Push the changes to GitHub: `git push -u origin master`
